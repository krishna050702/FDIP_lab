Canny Edge

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../assets/tiger.jpg', 0)

assert img is not None, 'file cannot be null'

edges = cv2.Canny(img, 100, 200)

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original img')
plt.xticks([])
plt.yticks([])

plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detected img')
plt.xticks([])
plt.yticks([])
plt.show()
````

Horizontal Lines

````python
import cv2
import numpy as np

img = cv2.imread('../assets/lines.png', 0)

# binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

kernel = np.ones((2, 19), np.uint8)

horizontalLines = cv2.erode(img, kernel, iterations=1)
# cv2.imshow('vertical', horizontalLines)

# cv2.imshow('horizontal', horizontalLines)
cv2.imwrite('horizontalLines.png', horizontalLines)

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original img')
plt.xticks([])
plt.yticks([])

plt.subplot(122)
plt.imshow(horizontalLines, cmap='gray')
plt.title('Horizontal Lines from given image')
plt.xticks([])
plt.yticks([])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
````

Vertical Lines

````python
import cv2
import numpy as np

img = cv2.imread('../assets/lines.png', 0)

# binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
kernel = np.ones((11, 3), np.uint8)
verticalLines = cv2.erode(img, kernel, iterations=1)
# cv2.imshow('vertical', verticalLines)

cv2.imwrite('../assets/verticalLines.png', verticalLines)

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original img')
plt.xticks([])
plt.yticks([])

plt.subplot(122)
plt.imshow(verticalLines, cmap='gray')
plt.title('Vertical Lines from given image')
plt.xticks([])
plt.yticks([])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
````

Hough Lines

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../assets/Lenna.png', 0)

assert img is not None, 'file cannot be null'

edges = cv2.Canny(img, 100, 150)
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, 30, maxLineGap=50)

for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1, y1), (x2, y2), (0, 0, 128), 1)

# cv2.imshow('line edges', edges)
# cv2.imshow('line detected', img)

cv2.imwrite('../assets/lineEdges.png', edges)
cv2.imwrite('../assets/lineDetected.png', img)

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Lines detected')
plt.xticks([])
plt.yticks([])

plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.title('Edges detected from given image')
plt.xticks([])
plt.yticks([])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
````
